```python
import seaborn as sns
import matplotlib.pyplot as plt

def get_barplot(column_name, data):
    counts = data[column_name].value_counts().sort_values()

    plt.figure(figsize=(10, 6))

    sns.barplot(x=counts.values, y=counts.index, color = '#4499F8')

    for index, value in enumerate(counts.values):
        plt.text(value, index, str(value), va='center')
        
    plt.title('Number of Observations per '+column_name)
    plt.xlabel('Count')
    plt.ylabel(column_name)

    plt.show()

get_barplot('brand', data)
```

```python
currency_rate = 0.012  # conversion rate from Indian rupees to USD
```

```python
# Currency change
data_USD = data
df = pd.DataFrame(data_USD)
df['Price'] = df['Price'] * currency_rate
df['Price'] = df['Price'].round(2)
```

```python
df.rename(columns={'Price': 'Price_USD'}, inplace=True)
data = data_USD
```

```python
y_min = data['Price_USD'].min() - 100
y_max = data['Price_USD'].max() + 100
```

```python
plt.figure(figsize=(40, 60))
i = 1
for val in brand_values:
    filtered_df = df[df['brand'] == val]
    plot = plt.subplot(math.ceil(len(brand_values)/7), 7, i)
    sns.boxplot(y=filtered_df['Price_USD'], color='lightblue')
    plot.set_xlabel(val, weight='bold', size=20)
    plt.ylim(y_min, y_max)
    i += 1

plt.suptitle('Boxplots of Price for different Brands', fontsize=40)
plt.show()
```
