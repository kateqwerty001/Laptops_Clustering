```python
import matplotlib.pyplot as plt
import seaborn as sns

for column in ['Price', 'Rating', 'num_cores', 'num_threads', 'ram_memory', 'primary_storage_capacity', 'secondary_storage_capacity', 'display_size', 'resolution_width', 'resolution_height']:
    plt.title(column)
    sns.boxplot(data[column])
    plt.show()
```
