```python
import seaborn as sns
import matplotlib.pyplot as plt

def get_barplot(column_name, data):
    counts = data[column_name].value_counts().sort_values()

    plt.figure(figsize=(10, 6))

    sns.barplot(x=counts.values, y=counts.index, color = '#4499F8')

    for index, value in enumerate(counts.values):
        plt.text(value, index, str(value), va='center')
        
    plt.title('Number of Observations per '+column_name)
    plt.xlabel('Count')
    plt.ylabel(column_name)

    plt.show()

get_barplot('brand', data)
```

#### What to do with outliers

```python
for column in ['Price', 'Rating', 'num_cores', 'num_threads', 'ram_memory', 'primary_storage_capacity', 'secondary_storage_capacity', 'display_size', 'resolution_width', 'resolution_height']:
    plt.title(column)
    sns.boxplot(data[column], color='lightblue')
    plt.show()
```

There are some outliers, but we have checked the prices of the most expensive laptops, and we see that these prices are real, so we decided to leave this outliers, because they are real data. 

```python
df = pd.DataFrame(data)

brand_counts = df['brand'].value_counts()
brand_top_values = np.array(brand_counts[brand_counts >= 5].index.tolist())
processor_brand_values = pd.unique(data['processor_brand'])
primary_storage_type_values = pd.unique(data['primary_storage_type'])
gpu_brand_values = pd.unique(data['gpu_brand'])
OS_counts = df['OS'].value_counts()
OS_top_values = np.array(OS_counts[OS_counts > 2].index.tolist())
```

```python
y_min = data['Price'].min() - 100
y_max = data['Price'].max() + 100
```

```python
plt.figure(figsize=(40, 10))
i = 1
for val in brand_top_values:
    filtered_df = df[df['brand'] == val]
    plot = plt.subplot(1, 11, i)
    sns.boxplot(y=filtered_df['Price'], color='lightblue')
    plot.set_xlabel(val, weight='bold', size=20)
    plt.ylim(y_min, y_max)
    i += 1

plt.suptitle('Boxplots of Price for the most popular Brands', fontsize=40)
plt.show()
```
